```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
```

```python
uber_df = pd.read_csv("uber-raw-data-sep14.csv")

uber_df.head(5)
```

```python
#display the last 5 records
uber_df.tail()
```

```python
uber_df.shape
```

```python
#understanhd the dataset properties
uber_df.info()
```

```python
# Let's break the date/time column to 'Day', 'Hour' & 'Weekday'
```

```python
# Change the 'Date, time' column's data type from string to datetime
uber_df['Date/Time'] = pd.to_datetime(uber_df['Date/Time'])
```

```python
# Change the 'Date, time' column's data type from string to datetime
uber_df['Day'] = uber_df['Date/Time'].apply(lambda x: x.day)
uber_df['Hour'] = uber_df['Date/Time'].apply(lambda x: x.hour)
uber_df['Weekday'] = uber_df['Date/Time'].apply(lambda x: x.weekday())
uber_df.head(5)
```

```python
#Now we can check the density of rides according to days,hours and weekdays
```

```python
#Visualize the density of rides per day of the month 
fig, ax = plt.subplots(figsize = (12,6))
plt.hist(uber_df.Day, width = 0.6,bins = 30)
plt.title("Density of trips per Day",fontsize = 16)
plt.xlabel('Day',fontsize = 14)
plt.ylabel('Density of rides',fontsize = 14) 
```

```python
#From the above plot we can notice that the highest number of rides are during working days (Monday to Friday),while the least number of rides in weekends.
```

```python
#Visualize the density of rides per weekday
fig,ax = plt.subplots(figsize = (12,6))
plt.hist(uber_df.Weekday,width = 0.6, range = (0,6.5), bins = 7, color = "purple")
plt.title("Density of trips per weekday", fontsize = 16)
plt.xlabel("Weekday", fontsize = 14)
plt.ylabel("Density of rides", fontsize = 14)
```

```python
#### The busiest day for Uber is Monday of the Week. On the other hand, the latest number of rides is on Saturday. 
```

```python
#Visuilaize the density of rides per hour.
fig,ax = plt.subplots(figsize = (12,6))
plt.hist(uber_df.Hour, width = 0.6, bins =24,color = "red")
plt.title("Density of trips per Hour",fontsize = 16)
plt.xlabel("Hour",fontsize = 14)
plt.ylabel("Density of rides", fontsize = 14)
```

```python
####The number of rides shows a gradual decrease from 1 AM to 4 PM, followed by an increase from 5 am onwards, reaching its peak at 6 pm. 
```

```python
#Visuilize the density of rides per location 
fig,ax = plt.subplots(figsize = (12,6))
x = uber_df.Lon
y = uber_df.Lat
plt.scatter(x, y, color = "orange")
plt.title("Density of trips per hour", fontsize = 16)
plt.xlabel("Hour",fontsize = 14)
plt.ylabel("Density of rides",fontsize = 14)
```
